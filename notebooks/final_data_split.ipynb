```python
## OBB->aa
import os
import glob

# Constants
IMG_WIDTH = 128
IMG_HEIGHT = 128

# Paths
base_label_path = "/home/rishabh.mondal/Brick-Kilns-project/ijcai_2025_kilns/data/processed_data/sentinel/stratified_split/labels"
aa_label_path = os.path.join(os.path.dirname(base_label_path), "aa_labels")
os.makedirs(aa_label_path, exist_ok=True)

def convert_obb_to_aabb(coords):
    xs = coords[::2]
    ys = coords[1::2]
    x_min, x_max = min(xs), max(xs)
    y_min, y_max = min(ys), max(ys)

    x_center = (x_min + x_max) / 2.0
    y_center = (y_min + y_max) / 2.0
    width = x_max - x_min
    height = y_max - y_min

    if width < 0 or height < 0:
        raise ValueError("Width and height must be positive values.")

    return [x_center, y_center, width, height]

# Process each label file
label_files = glob.glob(os.path.join(base_label_path, "*.txt"))

for label_file in label_files:
    with open(label_file, 'r') as f:
        lines = f.readlines()

    new_lines = []
    for line in lines:
        parts = line.strip().split()
        if len(parts) != 9:
            print(f"Skipping malformed line in {label_file}: {line}")
            continue
        class_id = parts[0]
        coords = list(map(float, parts[1:]))
        try:
            xc, yc, w, h = convert_obb_to_aabb(coords)
        except Exception as e:
            print(f"Error in {label_file}: {e}")
            continue

        new_lines.append(f"{class_id} {xc:.6f} {yc:.6f} {w:.6f} {h:.6f}")

    # Write to new file
    out_path = os.path.join(aa_label_path, os.path.basename(label_file))
    with open(out_path, 'w') as out_f:
        out_f.write('\n'.join(new_lines))

print(f"✅ Converted {len(label_files)} files and saved to: {aa_label_path}")
```

```python
## OBB -> DOTA Format
import os
from tqdm import tqdm

# Mapping from class ID to class name
class_mapping = {0: "CFCBK", 1: "FCBK", 2: "Zigzag"}

def convert_to_dota_format(input_path, output_path, image_width, image_height):
    """
    Convert a single label file from normalized format (YOLO-OBB style)
    to DOTA format.
    """
    with open(input_path, "r") as infile, open(output_path, "w") as outfile:
        for line in infile:
            values = line.strip().split()
            if len(values) < 9:
                continue  # Skip incomplete lines

            class_id = int(values[0])
            coordinates = list(map(float, values[1:9]))

            # Denormalize coordinates
            denormalized_coords = [
                round(coordinates[i] * (image_width if i % 2 == 0 else image_height), 2)
                for i in range(8)
            ]

            class_name = class_mapping.get(class_id, "UNKNOWN")
            # Format: x1 y1 x2 y2 x3 y3 x4 y4 class_name difficult
            dota_line = " ".join(map(str, denormalized_coords)) + f" {class_name} 0\n"
            outfile.write(dota_line)

def process_folder(input_folder, output_folder, image_width, image_height):
    """
    Process all `.txt` label files in the input folder and convert to DOTA format.
    """
    if not os.path.exists(input_folder):
        raise FileNotFoundError(f"Input folder '{input_folder}' not found")
    
    os.makedirs(output_folder, exist_ok=True)
    label_files = sorted([f for f in os.listdir(input_folder) if f.endswith(".txt")])

    for file_name in tqdm(label_files, desc="Converting to DOTA format", unit="file"):
        input_path = os.path.join(input_folder, file_name)
        output_path = os.path.join(output_folder, file_name)
        convert_to_dota_format(input_path, output_path, image_width, image_height)

# === Main Call ===
if __name__ == "__main__":
    input_folder = "/home/rishabh.mondal/Brick-Kilns-project/ijcai_2025_kilns/data/processed_data/sentinel/stratified_split/labels"
    output_folder = os.path.join(os.path.dirname(input_folder), "dota_labels")
    image_width = 128
    image_height = 128

    process_folder(input_folder, output_folder, image_width, image_height)

    print(f"\n✅ All label files converted and saved to: {output_folder}")
```

```python
#OBB -> COCO Format
import os
import json
import glob

IMG_WIDTH = 128
IMG_HEIGHT = 128

# Path configuration
image_dir = "/home/rishabh.mondal/Brick-Kilns-project/ijcai_2025_kilns/data/processed_data/sentinel/stratified_split/images"
label_dir = "/home/rishabh.mondal/Brick-Kilns-project/ijcai_2025_kilns/data/processed_data/sentinel/stratified_split/labels"
output_json = "/home/rishabh.mondal/Brick-Kilns-project/ijcai_2025_kilns/data/processed_data/sentinel/stratified_split/coco_format_labels.json"

# Category info
class_mapping = {0: "CFCBK", 1: "FCBK", 2: "Zigzag"}
categories = [{"id": k, "name": v, "supercategory": "kiln"} for k, v in class_mapping.items()]

images = []
annotations = []
ann_id = 1  # Annotation ID starts at 1

label_files = sorted(glob.glob(os.path.join(label_dir, "*.txt")))

for img_id, label_path in enumerate(label_files, 1):
    base_name = os.path.splitext(os.path.basename(label_path))[0]
    image_filename = base_name + ".png"
    image_path = os.path.join(image_dir, image_filename)

    # Register image info
    images.append({
        "id": img_id,
        "file_name": image_filename,
        "width": IMG_WIDTH,
        "height": IMG_HEIGHT
    })

    # Read and convert each label line
    with open(label_path, "r") as f:
        for line in f:
            parts = line.strip().split()
            if len(parts) != 9:
                continue  # skip invalid lines
            class_id = int(parts[0])
            coords = list(map(float, parts[1:9]))

            # Denormalize polygon
            poly = [
                round(coords[i] * (IMG_WIDTH if i % 2 == 0 else IMG_HEIGHT), 2)
                for i in range(8)
            ]

            # Bounding box from polygon (xmin, ymin, width, height)
            xs = poly[0::2]
            ys = poly[1::2]
            xmin = min(xs)
            xmax = max(xs)
            ymin = min(ys)
            ymax = max(ys)
            width = xmax - xmin
            height = ymax - ymin

            # COCO annotation
            annotations.append({
                "id": ann_id,
                "image_id": img_id,
                "category_id": class_id,
                "bbox": [xmin, ymin, width, height],
                "area": round(width * height, 2),
                "segmentation": [poly],
                "iscrowd": 0
            })
            ann_id += 1

# Final COCO structure
coco_dict = {
    "images": images,
    "annotations": annotations,
    "categories": categories
}

# Save to JSON
with open(output_json, "w") as f:
    json.dump(coco_dict, f, indent=4)

print(f"\n✅ COCO-format annotations saved to: {output_json}")
```

```python
import os
import shutil
import json
from sklearn.model_selection import StratifiedShuffleSplit
from collections import Counter

# ==== Base paths ====
base_path = "/home/rishabh.mondal/Brick-Kilns-project/ijcai_2025_kilns/data/processed_data/sentinel/stratified_split"
images_path = os.path.join(base_path, "images")
labels_path = os.path.join(base_path, "labels")
aa_labels_path = os.path.join(base_path, "aa_labels")
dota_labels_path = os.path.join(base_path, "dota_format_labels")
coco_path = os.path.join(base_path, "coco_format_labels.json")

# ==== Output split directories ====
splits = ['train', 'val', 'test']
for split in splits:
    os.makedirs(os.path.join(base_path, split, "images"), exist_ok=True)
    os.makedirs(os.path.join(base_path, split, "labels"), exist_ok=True)
    os.makedirs(os.path.join(base_path, split, "aa_labels"), exist_ok=True)
    os.makedirs(os.path.join(base_path, split, "dota_format_labels"), exist_ok=True)

# ==== Step 1: Collect image file names and dominant class ====
image_files = []
class_ids = []

for label_file in os.listdir(labels_path):
    if not label_file.endswith('.txt'):
        continue

    label_fp = os.path.join(labels_path, label_file)
    try:
        with open(label_fp, 'r') as f:
            lines = f.readlines()
            if not lines:
                continue
            labels = [int(line.strip().split()[0]) for line in lines]
            most_common_class = Counter(labels).most_common(1)[0][0]
            image_name = os.path.splitext(label_file)[0] + ".png"
            if os.path.exists(os.path.join(images_path, image_name)):
                image_files.append(os.path.splitext(label_file)[0])
                class_ids.append(most_common_class)
    except Exception as e:
        print(f"⚠️ Error reading {label_file}: {e}")

# ==== Step 2: Stratified split (60:20:20) ====
splitter1 = StratifiedShuffleSplit(n_splits=1, test_size=0.4, random_state=42)
train_idx, temp_idx = next(splitter1.split(image_files, class_ids))

temp_files = [image_files[i] for i in temp_idx]
temp_classes = [class_ids[i] for i in temp_idx]

splitter2 = StratifiedShuffleSplit(n_splits=1, test_size=0.5, random_state=42)
val_idx, test_idx = next(splitter2.split(temp_files, temp_classes))

train_files = [image_files[i] for i in train_idx]
val_files = [temp_files[i] for i in val_idx]
test_files = [temp_files[i] for i in test_idx]

# ==== Step 3: Copy files for image, labels, aa_labels, dota_format_labels ====
def copy_all_formats(files, split):
    for file in files:
        for ext in ['.jpg', '.png']:
            src_img = os.path.join(images_path, file + ext)
            if os.path.exists(src_img):
                break
        else:
            print(f"⚠️ Image file missing for {file}")
            continue

        # Copy image
        shutil.copy(src_img, os.path.join(base_path, split, "images", os.path.basename(src_img)))

        # Copy label formats
        for lbl_dir, subfolder in [
            (labels_path, "labels"),
            (aa_labels_path, "aa_labels"),
            (dota_labels_path, "dota_format_labels"),
        ]:
            src_lbl = os.path.join(lbl_dir, file + ".txt")
            dst_lbl = os.path.join(base_path, split, subfolder, file + ".txt")
            if os.path.exists(src_lbl):
                shutil.copy(src_lbl, dst_lbl)
            else:
                print(f"⚠️ Missing {subfolder} for {file}")

copy_all_formats(train_files, "train")
copy_all_formats(val_files, "val")
copy_all_formats(test_files, "test")

# ==== Step 4: Split COCO format JSON ====
with open(coco_path, "r") as f:
    coco_data = json.load(f)

filename_to_split = {
    fname + ".png": split
    for split, files in zip(["train", "val", "test"], [train_files, val_files, test_files])
    for fname in files
}

split_coco = {"train": [], "val": [], "test": []}
image_id_to_split = {}

# Split images
for image in coco_data["images"]:
    split = filename_to_split.get(image["file_name"])
    if split:
        image_id_to_split[image["id"]] = split
        split_coco[split].append(image)

# Split annotations
annotations_by_split = {"train": [], "val": [], "test": []}
for ann in coco_data["annotations"]:
    split = image_id_to_split.get(ann["image_id"])
    if split:
        annotations_by_split[split].append(ann)

# Save split JSON files
for split in ["train", "val", "test"]:
    output_json = {
        "images": split_coco[split],
        "annotations": annotations_by_split[split],
        "categories": coco_data["categories"]
    }
    out_path = os.path.join(base_path, f"coco_{split}.json")
    with open(out_path, "w") as f:
        json.dump(output_json, f, indent=4)
    print(f"✅ COCO {split} saved: {out_path}")

# ==== Final Summary ====
print(f"\n✅ Stratified split complete:")
print(f"  Train: {len(train_files)}")
print(f"  Val:   {len(val_files)}")
print(f"  Test:  {len(test_files)}")
```

```python
def count_classes_in_split(split_dir):
    label_dir = os.path.join(base_path, split_dir, "labels")
    class_counter = Counter()

    for label_file in os.listdir(label_dir):
        if not label_file.endswith(".txt"):
            continue
        label_path = os.path.join(label_dir, label_file)
        with open(label_path, "r") as f:
            for line in f:
                parts = line.strip().split()
                if parts:
                    class_id = int(parts[0])
                    class_counter[class_id] += 1
    return class_counter

print("\n📊 Class distribution per split:")
for split in ["train", "val", "test"]:
    counter = count_classes_in_split(split)
    print(f"  {split.capitalize()}:")
    for cls, count in sorted(counter.items()):
        print(f"    Class {cls}: {count}")
```
