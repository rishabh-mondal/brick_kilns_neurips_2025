```python
import os
import numpy as np
import pandas as pd
import geopandas as gpd
from shapely.geometry import Polygon

patch_size_pix = 128
patch_size_m = patch_size_pix * 10

def get_patch_degree_size(lat_center, patch_size_m=1280):
    deg_per_meter_lat = 1 / 111320
    deg_per_meter_lon = 1 / (111320 * np.cos(np.deg2rad(lat_center)))
    patch_height_deg = patch_size_m * deg_per_meter_lat
    patch_width_deg  = patch_size_m * deg_per_meter_lon
    return patch_height_deg, patch_width_deg

def get_patch_bounds(lat_center, lon_center, patch_height_deg, patch_width_deg):
    min_lon = lon_center - patch_width_deg / 2
    max_lon = lon_center + patch_width_deg / 2
    min_lat = lat_center - patch_height_deg / 2
    max_lat = lat_center + patch_height_deg / 2
    return min_lon, min_lat, max_lon, max_lat

def pixel_to_latlon(x, y, min_lon, min_lat, max_lon, max_lat, img_w=128, img_h=128):
    # Map (0,0) to (min_lon, max_lat), (img_w-1, img_h-1) to (max_lon, min_lat)
    lon = min_lon + (x / (img_w - 1)) * (max_lon - min_lon)
    lat = max_lat - (y / (img_h - 1)) * (max_lat - min_lat)
    return lon, lat

labels_dir = "/home/rishabh.mondal/Brick-Kilns-project/ijcai_2025_kilns/data/processed_data/sentinel/stratified_split/test/labels"
img_files = [f for f in os.listdir(labels_dir) if f.endswith('.txt')]

records = []
for idx, row in df.iterrows():
    lat_center = row['actual_lat_center']
    lon_center = row['actual_lon_center']
    patch_id = row['patch_id']
    label_path = os.path.join(labels_dir, f"{patch_id}.txt")
    if not os.path.exists(label_path) or pd.isna(lat_center) or pd.isna(lon_center):
        continue
    patch_height_deg, patch_width_deg = get_patch_degree_size(lat_center)
    min_lon, min_lat, max_lon, max_lat = get_patch_bounds(lat_center, lon_center, patch_height_deg, patch_width_deg)
    with open(label_path) as f:
        for line in f:
            vals = list(map(float, line.strip().split()))
            cls = int(vals[0])
            points = np.array(vals[1:]).reshape(4, 2)
            # Use (img_w-1), (img_h-1) for correct pixel index mapping
            points_pix = points * (patch_size_pix - 1)
            coords = [pixel_to_latlon(x, y, min_lon, min_lat, max_lon, max_lat, patch_size_pix, patch_size_pix) for x, y in points_pix]
            geom = Polygon(coords)
            records.append({
                'patch_id': patch_id,
                'class': cls,
                'geometry': geom
            })

labels_gdf = gpd.GeoDataFrame(records, crs="EPSG:4326")
labels_gdf.to_file("labels_polygons.geojson", driver="GeoJSON")
print(labels_gdf.head())
print(f"Total label polygons: {len(labels_gdf)}")
```

```python
def yolo_obb_dir_to_gdf(label_dir, patch_df, with_conf=False):
    records = []
    for idx, row in patch_df.iterrows():
        lat_center = row['actual_lat_center']
        lon_center = row['actual_lon_center']
        patch_id = row['patch_id']
        label_path = os.path.join(label_dir, f"{patch_id}.txt")
        if not os.path.exists(label_path) or pd.isna(lat_center) or pd.isna(lon_center):
            continue
        patch_height_deg, patch_width_deg = get_patch_degree_size(lat_center)
        min_lon, min_lat, max_lon, max_lat = get_patch_bounds(lat_center, lon_center, patch_height_deg, patch_width_deg)
        with open(label_path) as f:
            for line in f:
                vals = list(map(float, line.strip().split()))
                if with_conf:
                    cls = int(vals[0])
                    conf = float(vals[1])
                    points = np.array(vals[2:]).reshape(4, 2)
                else:
                    cls = int(vals[0])
                    conf = None
                    points = np.array(vals[1:]).reshape(4, 2)
                points_pix = points * (patch_size_pix - 1)
                coords = [pixel_to_latlon(x, y, min_lon, min_lat, max_lon, max_lat, patch_size_pix, patch_size_pix) for x, y in points_pix]
                geom = Polygon(coords)
                rec = {
                    'patch_id': patch_id,
                    'class': cls,
                    'geometry': geom
                }
                if with_conf:
                    rec['confidence'] = conf
                records.append(rec)
    gdf = gpd.GeoDataFrame(records, crs="EPSG:4326")
    return gdf
```

```python
import geopandas as gpd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# === Load data ===
pred_gdf = gpd.read_file('predictions.geojson')  # must have 'confidence'
gt_gdf = gpd.read_file('ground_truth.geojson')

# === Reproject to meters ===
pred_gdf = pred_gdf.to_crs('EPSG:3857')
gt_gdf = gt_gdf.to_crs('EPSG:3857')

# === Fix invalid geometries ===
def make_valid(gdf):
    gdf['geometry'] = gdf['geometry'].buffer(0)
    gdf = gdf[gdf.is_valid]
    return gdf.reset_index(drop=True)

pred_gdf = make_valid(pred_gdf)
gt_gdf = make_valid(gt_gdf)

print(f"Valid predictions: {len(pred_gdf)}")
print(f"Valid ground truths: {len(gt_gdf)}")

# === Compute IoU for each prediction ===
results = []
for idx_pred, pred in pred_gdf.iterrows():
    max_iou = 0
    best_gt_idx = None
    for idx_gt, gt in gt_gdf.iterrows():
        inter = pred.geometry.intersection(gt.geometry)
        union = pred.geometry.union(gt.geometry)
        if not inter.is_empty and union.area > 0:
            iou = inter.area / union.area
            if iou > max_iou:
                max_iou = iou
                best_gt_idx = idx_gt
    results.append({
        'confidence': pred.get('confidence', 1.0),
        'iou': max_iou,
        'area_m2': pred.geometry.area,
        'matched_gt_idx': best_gt_idx
    })

results_df = pd.DataFrame(results)

# === Plot: Confidence vs IoU (area as color) ===
plt.figure(figsize=(8,6))
sc = plt.scatter(
    results_df['confidence'],
    results_df['iou'],
    c=results_df['area_m2'],
    cmap='viridis',
    alpha=0.7
)
plt.colorbar(sc, label='Detection area (m²)')
plt.xlabel('Model Confidence')
plt.ylabel('IoU with GT')
plt.title('Model Confidence vs. IoU (Area Color-coded)')
plt.grid(True)
plt.show()
```
